##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
# Descriptive statististics using describe method
df.describe()
```

```python
# Calculate Measures of Central Tendency and Dispersion
from scipy.stats import mode

# Selecting key numerical variables
key_var = ['Income', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Recency']

stat = pd.DataFrame(index=key_var)
stat['Mean'] = df[key_var].mean()
stat['Mode'] = df[key_var].mode().iloc[0]
stat['Median'] = df[key_var].median()
stat['Standard Deviation'] = df[key_var].std()
stat['Variance'] = df[key_var].var()

print(stat)
```

```python
# identify and remove outlier

# Calculate quartiles and IQR for each variable
Q1 = df[key_var].quantile(0.25)
Q3 = df[key_var].quantile(0.75)
IQR = Q3 - Q1

# Determine outliers using IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = (df[key_var] < lower_bound) | (df[key_var] > upper_bound)

# Print columns with outliers
columns_with_outliers = outliers.any()
print("Columns with outliers:")
print(columns_with_outliers[columns_with_outliers].index.tolist())

# Remove outliers
df_no_outliers = df[~outliers.any(axis=1)]

# Print summary after removing outliers
print("\nDataset shape before removing outliers:", df.shape)
print("Dataset shape after removing outliers:", df_no_outliers.shape)
```

```python
# after handling outliers
print("\nAfter handling outliers:")

# Calculate Measures of Central Tendency and Dispersion
stat = pd.DataFrame(index=key_var)
stat['Mean'] = df[key_var].mean()
stat['Mode'] = df[key_var].mode().iloc[0]
stat['Median'] = df[key_var].median()
stat['Standard Deviation'] = df[key_var].std()
stat['Variance'] = df[key_var].var()
print(stat)
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

Binomial Distribution:

Complain:

The Complain variable represents whether a customer complained in the last 2 years (1 if complained, 0 otherwise).

Response:

The Response variable represents whether a customer accepted the offer in the last campaign (1 if accepted, 0 otherwise).

```python
# Probability and expected value of complain
prob_complain = df['Complain'].mean()
expected_complainers = len(df) * prob_complain
print(f"Probability of a customer complaining: {round(prob_complain,2)}")
print(f"Expected number of complainers: {expected_complainers}")

# Probability and expected value of response
prob_response = df['Response'].mean()
expected_responders = len(df) * prob_response
print(f"Probability of a customer responding: {round(prob_response,2)}")
print(f"Expected number of responders: {expected_responders}")
```

Normal Distribution:


Income: Yearly household income of customers.

MntWines: Amount spent on wine products in the last 2 years.

MntMeatProducts: Amount spent on meat products in the last 2 years.

MntGoldProds: Amount spent on gold products in the last 2 years.

```python
from scipy.stats import norm
# Define the variables of interest
var_ = ['Income', 'MntWines', 'MntMeatProducts', 'MntGoldProds']

# Loop through each variable
for var in var_:
  # Calculate mean and standard deviation
  mean_var = df[var].mean()
  std_dev_var = df[var].std()
  normal_dist = norm(mean_var, std_dev_var)

  # Calculate the probability of a specific condition (e.g., above mean)
  probability_above_mean = 1 - normal_dist.cdf(mean_var)

  # Calculate the expected value
  expected_value = normal_dist.mean()

  # Print the results for the variable
  print(f"{var}:")
  print(f"  Probability of {var} above its mean: {probability_above_mean:.4f}")
  print(f"  Expected {var}: {expected_value:.2f}\n")
```

## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

**Deliverables**:

- **Key insights regarding relationships between variables and distinct customer segments based on behaviors.**

  

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# create new column age
df['Age'] = 2024 - df['Year_Birth']

# Convert categorical variables to numeric
df['Education'] = pd.factorize(df['Education'])[0]
df['Marital_Status'] = pd.factorize(df['Marital_Status'])[0]

# Calculate 'Total_Spending' by summing up various spending columns
df['Total_Spending'] = df['MntFishProducts'] + df['MntMeatProducts'] + df['MntFruits'] + df['MntSweetProducts'] + df['MntWines'] + df['MntGoldProds']

# Select relevant features for segmentation
features = ['Age', 'Income', 'Total_Spending', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Recency']

# Standardize features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

# Clustering (K-means)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Visualize cluster characteristics
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Total_Spending', hue='Cluster', palette='viridis', s=100, alpha=0.8)
plt.title('Customer Segmentation based on Age vs Total Spending')
plt.xlabel('Age')
plt.ylabel('Total Spending')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Correlation matrix
corr_matrix = df[features].corr()
plt.figure(figsize=(10,6 ))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Customer Segmentation based on Age and Total Spending:

Key insight:

Cluster Analysis and Spending Patterns

Cluster 0 (Purple): Low total spending across all ages, likely representing low-value or less engaged customers.

Cluster 1 (Blue): High total spending spread across ages 30-70, indicating core customers with consistent purchasing habits.

Cluster 2 (Green): Low spending, typically younger customers, potential future high-value customers.

Cluster 3 (Yellow): Moderate spending across various ages, representing occasional shoppers.

High Spenders: Mostly in Cluster 1 (Blue), spending over 1500, aged 30-70.

Low Spenders: Predominantly in Cluster 0 (Purple), spending below 500.

Outliers: A few customers with exceptionally high spending across different clusters.

Age is positively correlated with income, indicating that older individuals tend to have higher incomes.

Income shows a strong positive correlation with total spending, catalog purchases, and store purchases, suggesting that higher income leads to increased spending and more purchases via catalogs and in stores.

Total spending is also positively correlated with web purchases but negatively correlated with website visits, implying that higher spenders visit the website less frequently.
